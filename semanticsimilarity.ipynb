```python
from nltk.corpus import wordnet as wn
import pandas as pd
import regex as re
import nltk
import numpy as np
```

```python
from GraphEditDistance import GED as ged

import itertools

def normalizarGED(ged, path1, path2):
    
    Xmax = max([path1,path2])
    res = 1- (ged/(Xmax))
    return res
        
lst = pd.read_csv('semanticsimilarity/list.csv')
lst = lst.drop_duplicates()
res = []
for index, couple in lst.iterrows():
    g = ged()   
    g = g.get_GED(couple['M'], couple['F'])
    
    w1 = g.syn1
    w2 = g.syn2
    if (w1 and w1):
        path_sim = w1.path_similarity(w2)
        lch = w1.lch_similarity(w2)
        wup = w1.wup_similarity(w2)
        norm = normalizarGED(g.GED, g.path1_to_root, g.path2_to_root)
        
        row = {'w1':g.word1,'w2':g.word2,'path similarity': path_sim, 'lch': lch, 'wup': wup,'normalizada:':norm,
               'GED':g.GED, 'hyp':g.hyp, 'GED1':g.GED1,'GED2':g.GED2}
        res.append(row)
        
    else:
        continue
df = pd.DataFrame(res)
print(df)
#df.to_csv('semanticsimilarity/listresultsgender.csv')
```
